# Aggregation

```python
iris.mean()
```

```python
# count, mean,...., max
iris.describe() 
```

```python
iris.groupby('species').mean()
```

```python
iris.groupby(['species', 'sepal_width']).mean()
```

```python
import numpy as np

pivotTable = pd.pivot_table(iris, values=['sepal_width', 'petal_width'], index=['species'],
                     aggfunc={'sepal_width': np.mean,
                              'petal_width': np.mean})
```

```python
pivotTable
```

```python
pivotTable.memory_usage() # useful prior to analsying data for machine learning
```

```python
pd.crosstab(index=iris['species'], columns=iris['petal_width'], margins=True)
```
